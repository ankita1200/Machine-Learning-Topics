Ordinary Least Square Regression -
In ordinary least square regression we want to find the weights that minimizes the sum of least squares.

Screenshot 2024-11-04 at 8.38.07 PM.png

```python
import numpy as np
```

```python
def linearRegression(X, y):
  X = np.array(X)
  y = np.array(y).reshape(-1,1)
  X_transpose = X.T
  # theta is weights
  theta = np.linalg.inv(X_transpose.dot(X)).dot(X_transpose).dot(y)
  theta = np.round(theta, 4).flatten().tolist()
  return theta
```

```python
X = [[1,2],[1,3],[2,4],[2,5]]
y = [1,2,3,4]
linearRegression(X,y)
```
