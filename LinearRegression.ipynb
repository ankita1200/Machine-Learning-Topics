```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([1,2,3,4,5])
b = np.array([2,4,5,4,5])
```

```python
mean_a = np.mean(a)
mean_b = np.mean(b)
```

```python
x = np.array([1,2,3,4,5])
y = np.array([2,4,5,4,5])
```

```python
mean_x = np.mean(x)
mean_y = np.mean(y)
m = (np.sum((x-mean_x*(y-mean_y))))/np.sum((x-mean_x)**2)
```

```python
c = mean_y - m*mean_x
```

```python
print(f"Linear model: y= {m}*x+{c}")
```

```python
plt.scatter(x,y, color='red',marker='o',s=30)
y_pred = m*x + c
plt.plot(x,y_pred, color='green',)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y/y_pred')
```
