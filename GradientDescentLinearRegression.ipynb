```python
import numpy as np
```

```python
def cost(X,y,theta):
  n = len(X)
  predictions = X.dot(theta)
  cost = np.sum((predictions-y)**2)/n
  return cost
```

```python
def gradient_descent(X,y,theta,iterations,alpha):
  cost_history = np.zeros(iterations)
  theta_history = np.zeros((iterations, X.shape[1]))
  for i in range(iterations):
    predictions = X.dot(theta)
    theta = theta - alpha * (1/len(X))* X.T.dot(predictions-y)
    theta_history[i,:] = theta.T
    cost_history[i] = cost(X,y,theta)
  return theta, theta_history, cost_history
```

```python
# Applying gradient descent to linear regression y = m*x + c
X = 2 * np.random.rand(100,1)
# randn - for normal disttribution (mean=0 and sd = 1)
y = 4 + 3 * X + np.random.randn(100,1)
learning_rate = 0.01
iterations = 1000
# randomly initialized parameters
theta = np.random.randn(2,1)
# add bias parameter
X_b = np.c_[np.ones((100,1)),X]
theta, theta_history, cost_history = gradient_descent(X_b,y,theta, iterations, learning_rate)
```
